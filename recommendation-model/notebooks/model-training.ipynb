```python
import pandas as pd
import numpy as np
from sklearn.decomposition import TruncatedSVD
from scipy.sparse import csr_matrix
```

```python
df = pd.read_csv("../data/training-data.csv")
df.head()
```

```python
user_item_matrix = df.pivot_table(
    index='user_id',
    columns='trip_id',
    values='rating'
).fillna(0)

user_item_matrix
```

```python
matrix = csr_matrix(user_item_matrix.values)
```

```python
svd = TruncatedSVD(n_components=2)
latent_matrix = svd.fit_transform(matrix)

latent_matrix
```

```python
import pickle

with open("../models/svd_model.pkl", "wb") as f:
    pickle.dump(svd, f)
```

```python
import numpy as np

# Choose a user
user_id = 1

# Get user vector
user_vector = latent_matrix[user_id - 1]

# Calculate scores for all trips
scores = user_vector.dot(svd.components_)

# Sort trips by score
recommended_trip_indices = np.argsort(scores)[::-1]

print("Recommended Trip IDs:")
print(recommended_trip_indices + 1)  # +1 because trip IDs start from 1
```
